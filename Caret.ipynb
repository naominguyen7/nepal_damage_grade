```python
from pycaret.classification import *
```

```python
exp_clf101 = setup(data = df, target = 'damage_grade', session_id=123, fold=5)
```

```python
best_model = compare_models(sort='F1', 
                            include=['xgboost', 'catboost', 'lightgbm', 'rf', 'et'])
```

```python
tuned_xgb = tune_model(best_model, optimize='F1')
```

```python
final_xgb = finalize_model(tuned_xgb)
```

```python
test_df = pd.read_csv('test_values.csv')
test_preds = predict_model(final_xgb, data=test_df)
```

```python
test_preds
```

```python
test_preds = test_preds[['building_id', 'Label']]
test_preds = test_preds.rename(columns={'Label':'damage_grade'})
test_preds.to_csv('submission_pycaret_notune.csv', index=False)
test_preds.head()
```

# Pycaret setup with tune

```python
cat_features = [
    'land_surface_condition',
    'foundation_type',
    'roof_type',
    'ground_floor_type',
    'other_floor_type',
    'position',
    'plan_configuration',
    'legal_ownership_status',
    'geo_level_2_id',
    'geo_level_1_id',
    'geo_level_3_id'
]

exp_clf102 = setup(data = df, target = 'damage_grade', session_id=123,
                   fold=5,
                   profile=True,
                   ignore_features=['building_id'],
                   categorical_features=cat_features,
                   normalize = True, 
                   transformation = True, 
                   ignore_low_variance = True,
                   feature_selection=True,
                   feature_selection_method='boruta',
                   log_experiment = True, experiment_name = 'pycaret-tuned-nepal')
```
